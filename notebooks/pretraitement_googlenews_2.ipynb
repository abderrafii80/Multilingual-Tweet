## Etape 6 : Clustering avec LDA 
### Il s'agit d'un document non annotes

## Preparation des donnees

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=100)
X = tfidf_vectorizer.fit_transform(df_ar['title'])
```

## Application de LDA

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=4, random_state=42)  # Ajustez n_components selon vos besoins
lda.fit(X)

# Représentation des documents par leur distribution de topics
topic_distribution = lda.transform(X)
```

```python
print(X)
```

## Clustering des documents 

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(topic_distribution)

# Ajoutez les clusters au DataFrame
df_ar['cluster'] = clusters
```

## Analyse des clusters

```python
terms = tfidf_vectorizer.get_feature_names_out()
for i, topic in enumerate(lda.components_):
    print(f"Topic {i}: {', '.join([terms[i] for i in topic.argsort()[-10:]])}")
```
